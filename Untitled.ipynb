```python
label = train['PE']
train.drop(columns=['PE'],inplace=True)
#test.drop(columns=['PE'],inplace=True)
```

```python
def model(m,train,label,test,splitters,rnd_st):
    prediction=pd.DataFrame(np.zeros(shape=(test.shape[0],splitters)))
    l1=[]
    l2=[]
    splitted = KFold(n_splits=splitters,random_state=rnd_st)
    
    for i,(tr_index,te_index) in enumerate(splitted.split(train,label)):
        x_train, x_test = train.loc[tr_index,:], train.loc[te_index,:]
        y_train,y_test = label[tr_index], label[te_index]
        
        m.fit(x_train,y_train)
        tr_pred = m.predict(x_train)
        te_pred = m.predict(x_test)
        
        l1.append(np.sqrt(mean_squared_error(y_train,tr_pred)))
        l2.append(np.sqrt(mean_squared_error(y_test,te_pred)))
        print('Training loss: {} \t\t\t\t Validation Loss {}'.format(np.sqrt(mean_squared_error(y_train,tr_pred)),np.sqrt(mean_squared_error(y_test,te_pred))))
    #    pd.Series(m.feature_importances_,index=train.columns).sort_values(ascending=True).plot.barh()
        
        prediction.iloc[:,i] = m.predict(test)
    
        
    print(np.mean(l1),np.mean(l2))
    return prediction
```

```python
seed = 90059
```

```python
final = model(lgb.LGBMRegressor(random_state=seed,n_estimators=5000),train,label,test,15,seed)
```

```python
pred = np.mean(final,axis=1)
```

```python
sample['PE'] = pred
```
